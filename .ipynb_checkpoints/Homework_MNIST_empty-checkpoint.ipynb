The first result is to analize the changes in performance of the SVM as the number of available labelled examples increases. The data for the first problem that you will study is in files "x_01.csv" and "y_01.csv". This problem consists in to separate 0s from 1s.

It seems a reasonably easy classification in terms of the separability of the classes. Therefore, it should be easy to achieve good classification results with a training dataset of relative small size. 

You are now analizing how the results of the SVM improve as you get more and more labelled  data. In the next cell implement python code that peforms the following functionalities.

- Run a `for` loop for all the values in the np.array `v_num_labels`. Let's say in iteration `nn` of the loop  `num_data = v_num_labels[nn]`
    - construct a training set of `num_data` observations by sampling at random `num_data` rows of `X` and the corresponding elements of `Y` as true labels. In this iteration of the loop you are simulating you only have labels for these `num_data` observations, therefore you can only use these observations to fit the SVM.
    - train a SVM with this training subset of size `num_data`
    - evaluate this SVM with `X` and `Y` and store the result in the position `nn` of `acc_svm_random_label`. Here you are using the whole dataset, but just for evaluation and reporting purposes, not for fitting the model.
    
- Plot `acc_svm_random_label` vs `v_num_labels` and discuss the impact of the number of available labelled examples in the classification accuracy

---

**Note**

It could happen that, specially for the smallest values of `v_num_lables`, all the `ll` samples belong to the same class. In that case store a $0.5$ in the corresponding position of `acc_svm_random_label` since you would think all the observations belong to a same class and therefore only half of the observations in `X` will get a correct classification.

```python
# Load data
X = pd.read_csv('x_01.csv').to_numpy()
Y = pd.read_csv('y_01.csv').to_numpy().ravel()

# Define parameters
C = 100
gamma = 0.001

# Define the number of labelled samples
v_num_labels = np.array([2, 4, 8, 10, 20, 30, 40, 60, 80, 100, 200, 300, 400])

# Initialize array to store accuracy values (for each iteration varying the number of labels)
acc_svm_random_label = np.empty(len(v_num_labels))

# Loop through different numbers of labelled samples
for nn, num_data in enumerate(v_num_labels):
    # Randomly select num_data training examples
    idx = np.random.choice(len(X), size=num_data, replace=False)
    X_train = X[idx] #we create a train set with the randomly selected labels 
    Y_train = Y[idx]
    
    # Check if both classes are present in the sampled data
    unique_classes = np.unique(Y_train)
    if len(unique_classes) < 2:
        acc_svm_random_label[nn] = 0.5  # Set accuracy to 0.5 for random guessing
    else:
        # Train SVM with the selected training examples
        clf = svm.SVC(kernel='rbf', C=C, gamma=gamma)
        clf.fit(X_train, Y_train)
    
        # Evaluate the SVM using the entire dataset
        accuracy = clf.score(X, Y)
        acc_svm_random_label[nn] = accuracy

# Plot accuracy vs number of labels
plt.figure()
plt.plot(v_num_labels, acc_svm_random_label, label='SVM, labels got at random')
plt.legend()
plt.xlabel('Num labels')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.grid()
plt.show()
```

COMMENTS: This plot shows the relationship between the number of labeled samples used for training an SVM model and the resulting classification accuracy. The x-axis is on a logarithmic scale and represents the number of labeled samples, while the y-axis represents the accuracy of the classifier.

From the plot, it seems that:

With a very small number of labels (less than 10), the accuracy quickly increases. This suggests that even a small number of labeled examples can be quite informative for the SVM when distinguishing between the digits 0 and 1 in the MNIST dataset. Given that these two digits are quite distinct in terms of their shapes, it is possible for an SVM to achieve reasonable separation with few examples.
Beyond 10 labels, the improvement in accuracy starts to plateau, reaching close to 1.0. This indicates that adding more labeled examples yields diminishing returns in terms of accuracy gains. It‚Äôs likely that the SVM has already found a decision boundary that generalizes well to the entire dataset, and additional data doesn‚Äôt significantly alter this boundary.
The initial steep improvement and subsequent plateau is a common pattern in learning curves, reflecting that initially, every new example can significantly improve the model's understanding of the decision boundary. However, once the model has 'learned' enough about the data, new examples only provide incremental improvements, as the model is already performing well.

The fact that the model achieves near-perfect accuracy indicates that the SVM is a suitable model for this task with the MNIST data and that the digits 0 and 1 are linearly separable to a high degree in the feature space defined by the pixel values when using an RBF kernel with the specified C and gamma parameters.

To get a more robust statistical significance of the results, in the next cell run the above loop a number of `NSIM` times, storing the results in the 2D array `acc_svm_random_label`. Then plot the mean of `acc_svm_random_label` and one standard deviation

```python
NSIM = 100
acc_svm_random_label = np.empty((NSIM, len(v_num_labels)))  # each row is for each of the NSIM simulations

# Loop through different simulations
for sim in range(NSIM):
    # Loop through different numbers of labelled samples (same loop as before)
    for nn, num_data in enumerate(v_num_labels):
        # Randomly select num_data training examples
        idx = np.random.choice(len(X), size=num_data, replace=False)
        X_train = X[idx]
        Y_train = Y[idx]

        # Check if both classes are present in the sampled data
        unique_classes = np.unique(Y_train)
        if len(unique_classes) < 2:
            acc_svm_random_label[sim, nn] = 0.5  # Set accuracy to 0.5 for random guessing
        else:
            # Train SVM with the selected training examples
            clf = svm.SVC(kernel='rbf', C=C, gamma=gamma)
            clf.fit(X_train, Y_train)

            # Evaluate the SVM using the entire dataset
            accuracy = clf.score(X, Y)
            acc_svm_random_label[sim, nn] = accuracy

# Plot mean and standard deviation of accuracy
plt.figure()
plt.plot(v_num_labels, np.mean(acc_svm_random_label, axis=0), label='SVM, labels got at random', color='blue')
plt.fill_between(v_num_labels, np.percentile(acc_svm_random_label, 25, axis=0),
                 np.percentile(acc_svm_random_label, 75, axis=0),
                 color='blue', alpha=0.3)
plt.legend()
plt.xlabel('Num labels')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.grid()
plt.show()
```

# 2.- Spectral Clustering with one label per cluster

Clustering is an unsupervised learning paradigm. It means we can use clustering algorithms with unlabelled data. A way of doing a classification with a very constrained budget for labels could be the following. Imagine you can afford `num_data` labels:
- Run a clustering algorithm and split your data in `num_data` clusters. We (optimistically) assume each cluster will be homogeneous (all the data in a same cluster should be assigned to a same class)
- Now run a loop for all the clusters to classify the data of the samples
    - Select one of the samples in each cluster and ask for its label, call `y_cluster` this label
    - Use that label to classify all the observations that belong to this cluster.

In the cell below complete the python code of function `my_spectral_clustering`. This function implements this classification scheme using [SpectralClustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.SpectralClustering.html) as clustering method.

Now use the function `my_spectral_clustering` to find out the performance of this proposed classification method based in clustering with a reduced number of labels using the number of clusters indicated by `v_num_labels` and a value of `gamma=0.001` for the RBF kernel of the spectral clustering. 

```python
from sklearn.cluster import SpectralClustering
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

class oracle(object):
    """
    Simulates an oracle that provides labels for input observations. 
    """
    def __init__(self, x, y):
        self.dim = x.shape[1]
        self.knn = KNeighborsClassifier(n_neighbors=1).fit(x, y)
        
    def get_label(self, x):
        return self.knn.predict(x.reshape(-1, self.dim))
    
def my_spectral_clustering(x, my_oracle, n_clusters=2, gamma=1):
    """
    Implements classification scheme using Spectral Clustering and a very limited number of labels.
    """
    Nt = x.shape[0]
    # Fit Spectral Clustering
    spectral = SpectralClustering(n_clusters=n_clusters, affinity='rbf', gamma=gamma, assign_labels='kmeans', random_state=0)
    cluster_labels = spectral.fit_predict(x)
    
    predictions = np.empty(Nt) # Array to fill with the predicted class for all instances in x
    
    for cc in range(n_clusters):
        # Get the instances that are in cluster cc
        indices = np.where(cluster_labels == cc)[0]
        if len(indices) > 0:
            # Pick one of these instances at random and call it my_x
            random_index = np.random.choice(indices)
            my_x = x[random_index]
            
            # Get the label for my_x from the oracle
            y_cluster = my_oracle.get_label(my_x)
            
            # Fill the positions of the instances in cluster cc in the array predictions with y_cluster
            predictions[indices] = y_cluster
    
    return predictions
```

```python
v_num_labels = np.array([2,4,8,10,20,30,40,60,80,100,200,300,400]) # number of labelled samples
acc_spectral_clustering = np.empty(len(v_num_labels))
gamma = 0.001
my_oracle = oracle(X,Y)
for ii,n_clusters in enumerate(v_num_labels):
    predictions = my_spectral_clustering(X, my_oracle, n_clusters=n_clusters, gamma=gamma)
    acc_spectral_clustering[ii] = np.mean(predictions == Y)
    
```

Plot the classification accuracy of the spectral clustering as a function of the number of available labels

```python
# YOUR CODE TO PRODUCE THE PLOT

plt.figure(figsize=(10, 6))
plt.plot(v_num_labels, acc_spectral_clustering, label='Spectral Clustering', color='green', marker='o')
plt.xlabel('Number of Labels')
plt.ylabel('Classification Accuracy')
plt.title('Classification Accuracy vs Number of Available Labels')
plt.xscale('log')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()

```

COMMENT: 

The plot shows that the classification accuracy is consistently high across different numbers of labels, all above 95%. This shows that spectral clustering is effective at grouping the data points into meaningful clusters that align well with the true class labels.

There is an initial increase in accuracy as the number of labels increases from 2 to around 30. This indicates that as more clusters are formed, they better capture the underlying class structure of the data, leading to more accurate classification when labeling one representative from each cluster.

Initially, increasing the number of clusters allows the spectral clustering algorithm to refine the boundaries between the classes. We actually only have 2 classes, 0 and 1; however, if we start with only 2 clusters, we are essentially hoping for a perfect separation between the classes. However, if there are subtle variations within the digits that could lead to overlap or misclassification, having more than 2 clusters allows the algorithm to capture these nuances. For example, different writing styles or orientations of the digits might lead to intermediate clusters that are not pure in terms of classes but help in better class separation.

After that, the accuracy seems to plateau with slight variability as the number of labels continues to increase. The plateau suggest that beyond a certain point, creating more clusters does not contribute to more accurate classification. This could happen when additional clusters start to over-segment the classes or split clusters that belong to the same class.  

There is a noticeable sharp drop in accuracy when the number of labels is 100, followed by a recovery. With the increase in the number of clusters, the probability of selecting a non-representative sample for labeling each cluster increases. If the randomly chosen sample is not central or typical of the cluster, the propagated label might not match the true labels of other samples in the cluster.

After the drop and subsequent recovery, the accuracy seems to stabilize again, though not quite reaching the previous peak. This stabilization suggests that with even more clusters, the method again starts to find a representative structure that approximates the true classes. However, the fact that it does not reach earlier highs in accuracy might indicate that there are diminishing returns to adding more clusters beyond a certain point.  As the number of clusters becomes very high, noise and outliers are increasingly likely to form their own clusters. This segregation can inadvertently correct some of the misclassification that occurred due to over-segmentation, leading to a rise in overall accuracy.

In conclusion, these results seem to support the idea that using spectral clustering with a very limited labeling budget can still result in high classification accuracy. However, there is a level of complexity or a "sweet spot" in terms of the number of clusters where the accuracy is maximized before it becomes counterproductive to add more clusters. The drop and recovery in the plot underscore the importance of choosing the right number of clusters for spectral clustering to be effective, as well as the risks associated with the strategy of labeling only one data point per cluster.

Explore the dependence of the classification accuracy of the spectral clustering with the values of `gamma`. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import SpectralClustering
from sklearn.metrics import accuracy_score

v_num_labels = np.array([2,4,8,10,20,30,40,60,80,100,200,300,400])
v_gamma = [1e-4, 1e-3, 1e-2, 1e-1]
acc_spectral_clustering = np.empty((len(v_gamma), len(v_num_labels)))

for ig, gamma in enumerate(v_gamma):
    for ii, n_clusters in enumerate(v_num_labels):
        predictions = my_spectral_clustering(X, my_oracle, n_clusters=n_clusters, gamma=gamma)
        acc_spectral_clustering[ig, ii] = accuracy_score(Y, predictions)

# Now plot the results
plt.figure()
for ig, gamma in enumerate(v_gamma):
    plt.plot(v_num_labels, acc_spectral_clustering[ig, :], label=f'Spectral Clustering, gamma {gamma:.4f}')
plt.legend()
plt.xlabel('Number of Labels')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.grid()
plt.show()
```


The parameter gamma in spectral clustering controls the width of the neighborhood in the similarity graph constructed by the RBF kernel. A high gamma value leads to a graph where only the nearest neighbors are strongly connected. On the contrary, a low gamma value creates a smoother graph where distant points can influence each other, potentially leading to fewer, larger clusters.

Interpretation of Each Gamma Line:

**Gamma = 0.1 (Red Line)**:
The high gamma value results in significant variability in classification accuracy. This suggests that the algorithm is sensitive to the local structure of the data, perhaps too much so. At this scale, the algorithm may create clusters that are too fine, potentially dividing what should be cohesive classes into multiple clusters. This results in the severe drops in accuracy seen, especially when the number of labels is moderate to high. The high gamma value may be capturing noise as structure, which is detrimental to the classification performance.

**Gamma = 0.01 (Green Line)**:
A tenfold decrease in gamma produces a more stable accuracy, though with some fluctuations. This intermediate gamma value seems to be less sensitive to noise and more capable of capturing the true underlying structure of the data. However, it still exhibits some variability, which suggests that while it is much better than a gamma of 0.1, it might still be creating clusters that do not perfectly align with the actual class divisions.

**Gamma = 0.001 (Orange Line)**:
The orange line represents a gamma that appears to balance well between the global and local structures, offering good performance across a range of label numbers. At this scale, the similarity graph is likely reflecting a more accurate global structure without being overly disrupted by local noise. This gamma value seems to allow the algorithm to separate the classes effectively while also being robust to the number of clusters specified.

**Gamma = 0.0001 (Blue Line)**:
The blue line, which corresponds to the lowest gamma, shows one of the highest and most consistent accuracy. This suggests that the broader kernel width, which allows for distant points to influence each other, is beneficial for this dataset. It may indicate that the true class structure in the data is relatively simple or that the classes are separated by wide margins. A broader neighborhood helps in clearly defining class boundaries without being misled by within-class variability or noise.

With a high gamma, the similarity graph becomes more local. Only points that are very close to each other are considered similar, leading to a graph with many small neighborhoods. When we ask the spectral clustering algorithm to find clusters in this graph, it tends to create more, smaller clusters. However, if we specify fewer clusters than the graph's inherent local structure suggests, the algorithm may force together points that are not truly similar, leading to poor performance. On the other hand, a low gamma value creates a similarity graph where even distant points can be considered part of the same neighborhood. The algorithm sees a "bigger picture" of the data. In this case, if the similarity graph suggests fewer, larger clusters and we specify too many, we may end up splitting cohesive groups unnecessarily, again degrading performance. The best performance is typically achieved when the number of clusters we specify matches the true number of distinct groups in the data, as represented by the similarity graph constructed with an appropriate gamma value.

# 3.1 SVM

```python
import numpy as np
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt


# Load data
Y = pd.read_csv("y_all_even_odd.csv").to_numpy().ravel()
X = pd.read_csv("x_all_even_odd.csv").to_numpy()

# Define parameters
C = 100
gamma = 0.001

# Define the number of labelled samples
v_num_labels = np.array([2, 4, 8, 10, 20, 30, 40, 60, 80, 100, 200, 300, 400])

# Initialize array to store accuracy values
acc_svm_random_label = np.empty(len(v_num_labels))

# Loop through different numbers of labelled samples
for nn, num_data in enumerate(v_num_labels):
    # Randomly select num_data training examples
    idx = np.random.choice(len(X), size=num_data, replace=False)
    X_train = X[idx]
    Y_train = Y[idx]
    
    # Check if both classes are present in the sampled data
    unique_classes = np.unique(Y_train)
    if len(unique_classes) < 2:
        acc_svm_random_label[nn] = 0.5  # Set accuracy to 0.5 for random guessing
    else:
        # Train SVM with the selected training examples
        clf = svm.SVC(kernel='rbf', C=C, gamma=gamma)
        clf.fit(X_train, Y_train)
    
        # Evaluate the SVM using the entire dataset
        accuracy = clf.score(X, Y)
        acc_svm_random_label[nn] = accuracy

# Plot accuracy vs number of labels
plt.figure()
plt.plot(v_num_labels, acc_svm_random_label, label='SVM, labels got at random')
plt.legend()
plt.xlabel('Num labels')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.grid()
plt.show()


```


When the number of labeled examples is smallest, the SVM's accuracy is not much better than random guessing. This suggests that with very few labeled points, the SVM cannot construct a reliable model of the data's underlying structure. As the number of labels increases, there is a sharp, almost linear increase in accuracy. This reflects that each additional labeled example significantly contributes to the SVM's ability to define the decision boundary. Given the more complex nature of the "even vs. odd" classification task compared to the "0 vs. 1" task, the SVM requires a substantial number of examples to capture the nuances necessary to differentiate between the classes.
As more labels are provided, the rate of increase in accuracy starts to slow down, and the curve begins to plateau, although it is still on an upward trend at the end of the plot. This behavior is typical of learning curves in machine learning, where the marginal benefit of additional data diminishes as the model becomes more refined.
If we were to compare this curve to the one obtained from the "0 vs. 1" classification, we might notice a couple of differences. First, the starting accuracy for the "even vs. odd" task is likely lower, reflecting the increased complexity of distinguishing between multiple pairs of digits as opposed to just two. Second, the rate at which accuracy increases with additional labels might be less steep in the "even vs. odd" case, again due to the task's greater complexity.
Despite the more challenging classification task, the plot suggests that given a sufficient number of labeled examples, the SVM can still achieve a high level of accuracy. # 3.2 SPECTRAL CLUSTER


```python
from sklearn.cluster import SpectralClustering
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

class oracle(object):
    """
    Simulates an oracle that provides labels for input observations. 
    """
    def __init__(self, x, y):
        self.dim = x.shape[1]
        self.knn = KNeighborsClassifier(n_neighbors=1).fit(x, y)
        
    def get_label(self, x):
        return self.knn.predict(x.reshape(-1, self.dim))
    
def my_spectral_clustering(x, my_oracle, n_clusters=2, gamma=1):
    """
    Implements classification scheme using Spectral Clustering and a very limited number of labels.
    """
    Nt = x.shape[0]
    # Fit Spectral Clustering
    spectral = SpectralClustering(n_clusters=n_clusters, affinity='rbf', gamma=gamma, assign_labels='kmeans', random_state=0)
    cluster_labels = spectral.fit_predict(x)
    
    predictions = np.empty(Nt) # Array to fill with the predicted class for all instances in x
    
    for cc in range(n_clusters):
        # Get the instances that are in cluster cc
        indices = np.where(cluster_labels == cc)[0]
        if len(indices) > 0:
            # Pick one of these instances at random and call it my_x
            random_index = np.random.choice(indices)
            my_x = x[random_index]
            
            # Get the label for my_x from the oracle
            y_cluster = my_oracle.get_label(my_x)
            
            # Fill the positions of the instances in cluster cc in the array predictions with y_cluster
            predictions[indices] = y_cluster
    
    return predictions
```

```python
v_num_labels = np.array([2,4,8,10,20,30,40,60,80,100,200,300,400]) # number of labelled samples
acc_spectral_clustering = np.empty(len(v_num_labels))
gamma = 0.001
my_oracle = oracle(X,Y)
for ii,n_clusters in enumerate(v_num_labels):
    predictions = my_spectral_clustering(X, my_oracle, n_clusters=n_clusters, gamma=gamma)
    acc_spectral_clustering[ii] = np.mean(predictions == Y)
```

Starting from a low accuracy that is just above random chance with very few labels, the accuracy climbs as more labels are provided, this can be explained by the fact that ff the data has inherent groupings that are more granular than just "even" versus "odd," initially increasing the number of clusters can help the algorithm to capture these subtleties. For instance, within the "even" numbers, digits like "0" and "8" might have similar pixel patterns that are distinct from "2" and "4", so more clusters can initially help to distinguish between these sub-groups better.

However, the plot also exhibits fluctuations, with peaks and valleys in accuracy as the number of clusters continues to increase. This "wiggly" behavior likely reflects the variability inherent in the process of assigning labels to clusters based on a single representative. Depending on which instances are selected as representatives, the labeling could be more or less representative of their respective clusters.

Despite the variability, the general trend suggests an improvement in accuracy with more clusters. This improvement implies that the additional clusters are beneficial in capturing the data's structure more accurately. Nevertheless, there is a notable increase in accuracy at the highest cluster counts, suggesting that the algorithm finds a more accurate clustering of the data that aligns well with the even-odd classes as it is provided with more clusters to work with. When we specify more clusters than there are true categories, the spectral clustering algorithm will still find a way to divide the data into the specified number of clusters. If by chance these additional clusters align well with some of the natural variations within the "even" and "odd" classes, then labeling a representative from each cluster can lead to more accurate classification, as subtle distinctions within the classes are being accounted for.

```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.cluster import SpectralClustering
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Load data
Y = pd.read_csv("y_all_even_odd.csv").to_numpy().ravel()
X = pd.read_csv("x_all_even_odd.csv").to_numpy()

# Define parameters
C = 100
gamma = 0.001

# Define the number of labelled samples and simulations
v_num_labels = np.array([2, 4, 8, 10, 20, 30, 40, 60, 80, 100, 200, 300, 400])
NSIM = 100

# Initialize array to store accuracy values for each simulation
acc_svm_spectral_label = np.empty((NSIM, len(v_num_labels)))

# Oracle class definition
class oracle:
    def __init__(self, x, y):
        self.dim = x.shape[1]
        self.knn = KNeighborsClassifier(n_neighbors=1).fit(x, y)
        
    def get_label(self, x):
        return self.knn.predict(x.reshape(-1, self.dim))

# Initialize the oracle with the dataset
my_oracle = oracle(X, Y)

# Loop through different simulations
for sim in range(NSIM):
    for nn, n_clusters in enumerate(v_num_labels):
        # Perform spectral clustering
        spectral = SpectralClustering(n_clusters=n_clusters, affinity='rbf', gamma=gamma, assign_labels='kmeans', random_state=sim)
        cluster_labels = spectral.fit_predict(X)

        # Prepare training data for SVM
        X_train = []
        Y_train = []
        
        # Select one sample per cluster to train the SVM
        for cc in range(n_clusters):
            indices = np.where(cluster_labels == cc)[0]
            if len(indices) > 0:
                random_index = np.random.choice(indices)
                X_train.append(X[random_index])
                Y_train.append(Y[random_index])
        
        X_train = np.array(X_train)
        Y_train = np.array(Y_train)
        
        # Check if there are at least two different classes
        if len(np.unique(Y_train)) > 1:
            # Train SVM with samples selected by spectral clustering
            clf = svm.SVC(kernel='rbf', C=C, gamma=gamma)
            clf.fit(X_train, Y_train)

            # Evaluate the SVM on the entire dataset
            accuracy = clf.score(X, Y)
            acc_svm_spectral_label[sim, nn] = accuracy
        else:
            # Not enough classes for a meaningful training
            acc_svm_spectral_label[sim, nn] = 0.5

# Calculate the mean and confidence interval for the accuracies
mean_accuracy = np.mean(acc_svm_spectral_label, axis=0)
confidence_interval = np.std(acc_svm_spectral_label, axis=0) * 1.96 / np.sqrt(NSIM)

# Plot mean accuracy and confidence interval
plt.figure(figsize=(10, 6))
plt.plot(v_num_labels, mean_accuracy, label='SVM, labels from spectral clustering', color='red')
plt.fill_between(v_num_labels, mean_accuracy - confidence_interval, mean_accuracy + confidence_interval, color='red', alpha=0.2)
plt.legend()
plt.xlabel('Number of Labels')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.title('Accuracy of SVM trained on labels from spectral clustering')
plt.grid()
plt.show()

```
